tf.summary used for setup and store tensorboard file

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
summary_writer = tf.summary.create_file_writer('lab7-logs-summary')
```

with summary_writer.as_default():
    tf.summary.(scalar | image | text | audio | histogram)(...) #設定寫入的記錄類型

tf.summary.scalar used for store value such as loss and metrics

```python
#在0-2pi 間產生100點
x = np.linspace(0, 2 * np.pi, 100)
#將100點帶入sin 函數中產生資料
data = np.sin(x)
with summary_writer.as_default():
    for i, y in enumerate(data):
        tf.summary.scalar('sin', y, step=i)
        #y-axis = value, x-axis = time
```

tf.summary.image is used for storing image

```python
def read_img(file):
    image_string = tf.io.read_file(file)
    #以影像格式解碼
    image_decode = tf.image.decode_image(image_string)
    return image_decode

img = read_img('image/airplane.png')
plt.imshow(img)
```

```python
with summary_writer.as_default(): #summary_writer as record file
    tf.summary.image("Airplane", [img], step=0) #save the image
```

```python
img_files = ['airplane_zoom.png', 'airplane_flip.png', 'airplane_color.png', 'airplane_rot.png',
             'airplane.png'] 

imgs = []
for file in img_files:
    imgs.append(read_img('image/'+file))
    
with summary_writer.as_default():
    tf.summary.image("Airplane Augmentation", imgs, max_outputs=5, step=0)
```

```python
with summary_writer.as_default():
    for i, img in enumerate(imgs):
        tf.summary.image("Save image each step", [img], step=i)
```

tf.summary.text used for saving text information

```python
# 建立一個陣列，裡面包含了對話記錄
texts = ["小明：Cubee小助理最近好想學深度學習的技術哦!", 
         "Cubee：這是當然的阿，這可現今最火的技術呢!", 
         "小明：那我該如何入門呢?", 
         "Cubee：推薦你一本書「輕鬆學會Google TensorFlow2.0深度學習」。", 
         "小明：這本書沒有深度學習經驗的人也能學會嗎?", 
         "Cubee：這是當然的，你只需要基礎Python能力就可以學會了!", 
         "小明：太好了那我要趕快去買了!"]

with summary_writer.as_default():
    for i, text in enumerate(texts):
        tf.summary.text("Chat record", text, step=i)
```

tf.summary.audio used for saving audio

```python
def read_audio(file):
    audio_string = tf.io.read_file(file)
    audio, fs = tf.audio.decode_wav(audio_string)
    # 因為tf.summary.audio要求輸入格式為[k(clips), t(frames), c(channels)]
    # 而解碼後的音訊只有[t(frames), c(channels)]，所以需要增加一個維度給音訊
    audio = tf.expand_dims(audio, axis=0)
    return audio, fs

audio, fs = read_audio('./audio/cat.wav')
```

```python
with summary_writer.as_default():
    tf.summary.audio('cat', audio, fs, step=0)
```

tf.summary.histogram used for storing data distribution function and graph

tf.keras.callbacks.TensorBoard only store weight distribution
tf.summary.histogram API store output distribution

```python
#建立一個常態分佈normal distribution 資料，資料共64筆，每筆有100個資料
data = tf.random.normal([64, 100], dtype=tf.float64)
with summary_writer.as_default():
    tf.summary.histogram('Normal distribution', data, step=0)
```

```python
with summary_writer.as_default():
    #儲存100組常態分佈，並且各個之間平均值都相差0.01
    for i, offset in enumerate(tf.range(0, 10, delta=0.1, dtype=tf.float64)):
        tf.summary.histogram('Normal distribution 2', data+offset, step=i)
```
